## Normalização de dados

### Definição e importância da normalização

#### Definição e importância da normalização

A normalização na preparação de dados contínuos é um processo que visa ajustar os valores dos dados para uma escala comum. Ela desempenha um papel fundamental na análise de dados e no aprendizado de máquina, pois elimina o viés introduzido por diferenças nas unidades de medida e magnitudes dos dados.

A definição da normalização consiste em transformar os dados de forma que eles estejam em uma escala comparável. Isso significa que os valores originais são redimensionados para um intervalo específico, geralmente entre 0 e 1, ou são transformados para terem média zero e desvio padrão um.

A importância da normalização reside em vários aspectos. Primeiramente, ela garante que todos os recursos ou variáveis tenham uma influência igual durante a análise ou o treinamento de modelos de aprendizado de máquina. Sem a normalização, características com escalas maiores podem dominar ou ter um impacto desproporcional em comparação com características com escalas menores.

Além disso, a normalização ajuda a evitar problemas relacionados a unidades de medida diferentes. Dados com unidades distintas podem distorcer a análise ou os modelos, pois as diferenças nas escalas podem levar a interpretações errôneas dos resultados. A normalização permite comparar e combinar características que estão em diferentes escalas, tornando-as comparáveis e facilitando a interpretação dos resultados.

Outro benefício da normalização é a capacidade de lidar com valores extremos. Em algumas técnicas de normalização, como a normalização min-max, os valores extremos são preservados, mantendo sua importância relativa nos dados. Isso é útil em cenários onde esses valores extremos têm significado especial ou podem fornecer informações valiosas.

Em resumo, a normalização na preparação de dados contínuos é crucial para garantir que as características sejam comparáveis e tratadas de forma equitativa. Ela permite que os algoritmos de aprendizado de máquina trabalhem com eficácia, melhorando o desempenho e a precisão dos modelos. Além disso, a normalização facilita a interpretação dos resultados, eliminando o viés introduzido por unidades de medida e magnitudes diferentes.

#### Objetivos da normalização

A normalização na preparação de dados contínuos tem como objetivo principal ajustar os valores dos dados para uma escala comum. Essa transformação busca atingir os seguintes objetivos:

##### Eliminação do viés introduzido por unidades de medida diferentes

Diferentes variáveis podem ter unidades de medida distintas, o que pode levar a interpretações errôneas ou enviesadas dos dados. A normalização permite remover essa disparidade, colocando todas as variáveis em uma escala comparável. Isso é especialmente importante em análises que envolvem a combinação ou comparação de múltiplas variáveis.

Um exemplo desse objetivo pode ser: suponha que uma pessoa esteja trabalhando em um projeto de análise de dados de saúde, no qual deseja investigar a relação entre o índice de massa corporal (IMC) e a pressão arterial de um grupo de indivíduos. Essa pessoa coletou dados de 100 participantes, medindo seu IMC em quilogramas por metro quadrado (kg/m²) e sua pressão arterial em milímetros de mercúrio (mmHg).

Sem a normalização, as diferenças nas unidades de medida podem distorcer a análise e as conclusões tiradas dos dados. O IMC, por exemplo, pode variar de 10 a 40 kg/m², enquanto a pressão arterial pode variar de 80 a 120 mmHg. Se os valores forem comparados diretamente sem normalização, a escala maior da pressão arterial pode dominar a análise, levando a conclusões equivocadas.

Para resolver esse problema, a pessoa decide normalizar os dados. Ela aplica a normalização min-max ao IMC e a normalização z-score à pressão arterial. A normalização min-max redimensiona os valores do IMC para uma escala entre 0 e 1, enquanto a normalização z-score transforma a pressão arterial para ter média zero e desvio padrão um.

Após a normalização, é possível comparar e combinar as duas variáveis de forma adequada. Agora, tanto o IMC quanto a pressão arterial estão em escalas comparáveis, eliminando o viés introduzido pelas unidades de medida diferentes. Isso permite que a pessoa investigue a relação entre essas variáveis de maneira equitativa e tire conclusões mais precisas.

Com a normalização, é possível identificar padrões claros, como uma correlação negativa entre o IMC e a pressão arterial, indicando que os indivíduos com IMC mais alto tendem a ter pressão arterial mais alta. Essa interpretação é possível porque a normalização nivelou as unidades de medida e tornou as variáveis comparáveis, evitando interpretações errôneas ou enviesadas dos dados.

Portanto, ao eliminar o viés introduzido por unidades de medida diferentes por meio da normalização, é possível realizar análises mais confiáveis e obter insights valiosos sobre as relações entre variáveis no projeto de análise de dados.


##### Igualdade de tratamento dos recursos

Ao normalizar os dados, todas as características ou recursos são tratados igualmente pelos algoritmos de análise ou modelos de aprendizado de máquina. Sem a normalização, recursos com escalas maiores podem dominar e ter um impacto desproporcional na análise ou nos resultados dos modelos. A normalização garante que todos os recursos tenham a mesma influência durante o processo de análise e modelagem, evitando assim o enviesamento dos resultados.

Por exemplo, em um estudo de mercado, um analista de dados está trabalhando em um projeto para prever o preço de venda de imóveis com base em várias características, como área, número de quartos, número de banheiros e idade da propriedade. Essas características são medidas em escalas diferentes: a área é medida em metros quadrados, o número de quartos e banheiros são valores inteiros, e a idade da propriedade é medida em anos.

Sem a normalização, as características com escalas maiores, como a área, podem ter um impacto desproporcional na análise e nos resultados dos modelos de aprendizado de máquina. Por exemplo, se a área variar de 50 a 200 metros quadrados, enquanto o número de quartos varia de 1 a 4, um modelo de aprendizado de máquina pode dar mais peso à área, considerando-a como o fator dominante para determinar o preço de venda dos imóveis.

Para evitar esse enviesamento e tratar todas as características igualmente, o analista decide normalizar os dados. Ele aplica a normalização min-max a todas as características, redimensionando os valores para uma escala entre 0 e 1. Dessa forma, todas as características estão em uma escala comparável, e nenhuma delas terá um impacto desproporcional durante a análise ou modelagem.

Com a normalização, todas as características, independentemente de sua escala original, têm a mesma influência na previsão do preço de venda dos imóveis. O modelo de aprendizado de máquina pode considerar todas as características de forma justa, sem favorecer aquelas com escalas maiores. Isso evita o enviesamento dos resultados e permite que o analista obtenha previsões mais precisas e confiáveis sobre os preços de venda dos imóveis.

Portanto, ao normalizar os dados, todas as características são tratadas igualmente pelos algoritmos de análise ou modelos de aprendizado de máquina. Essa equalização de influência garante uma análise justa e imparcial, evitando que recursos com escalas maiores dominem os resultados. A normalização contribui para um processo de análise e modelagem mais robusto e confiável.

##### Comparação da posição relativa dos valores

A normalização, em particular a normalização z-score, permite comparar a posição relativa dos valores em relação à média. Isso é útil para identificar valores atípicos (outliers) e entender como cada valor se comporta em relação à distribuição dos dados. A padronização facilita a identificação de valores que estão acima ou abaixo da média, ajudando a identificar padrões e anomalias nos dados.

Por exemplo, em um estudo de análise de dados de desempenho acadêmico, um pesquisador está investigando o desempenho dos estudantes em diferentes disciplinas. Para compreender melhor os padrões de desempenho dos estudantes em cada disciplina e identificar possíveis valores atípicos, o pesquisador decide aplicar a normalização z-score aos dados. A normalização z-score transforma cada valor em relação à média e ao desvio padrão da respectiva disciplina.

Em resumo, a normalização z-score permite que o pesquisador compare a posição relativa dos valores em relação à média e ao desvio padrão. Isso facilita a identificação de valores atípicos, tanto acima quanto abaixo da média, e ajuda a entender como cada valor se comporta em relação à distribuição dos dados. A padronização contribui para a identificação de padrões e anomalias nos dados, permitindo insights mais precisos sobre o desempenho dos estudantes em diferentes disciplinas.

Em resumo, os objetivos da normalização na preparação de dados contínuos são eliminar o viés introduzido por unidades de medida, garantir a igualdade de tratamento dos recursos, permitir a comparação da posição relativa dos valores, melhorar o desempenho e a precisão dos modelos, além de facilitar a interpretação dos resultados. Esses objetivos são essenciais para garantir uma análise robusta e confiável dos dados e para obter insights significativos a partir deles.

## Técnicas de normalização

### Normalização min-max (escala)

A técnica de normalização min-max é um método comum para transformar os valores de uma variável em uma escala específica, geralmente entre 0 e 1. Isso é feito por meio de uma fórmula simples que considera o valor mínimo e máximo da variável original.

A fórmula para o cálculo da normalização min-max é a seguinte:

$$X_{normalizado} = \frac{X−X_{min}}{X_{max}−X_{min}}$$

Onde:

- $X$ é o valor original da variável;
- $X_{min}$ é o valor mínimo observado na variável;
- $X_{max}$ é o valor máximo observado na variável;
- $X_{normalizado}$ é o valor normalizado da variável.

Para realizar o cálculo, o pesquisador subtrai o valor mínimo da variável original ($X_{min}$) do valor original ($X$). Em seguida, o resultado é dividido pela diferença entre o valor máximo da variável original ($X_{max}$) e o valor mínimo ($X_{min}$).

Dessa forma, o valor normalizado resultante ($X_{normalizado}$) estará dentro do intervalo de 0 a 1, onde 0 representa o valor mínimo observado e 1 representa o valor máximo observado na variável original. Os valores intermediários serão proporcionais em relação a essa escala.

Essa técnica de normalização é amplamente utilizada em análises de dados para trazer todas as variáveis para uma mesma escala comparável, eliminando as diferenças de magnitude entre os valores. Isso facilita a comparação e a interpretação dos dados, além de evitar que variáveis com escalas maiores dominem a análise ou os modelos de aprendizado de máquina.

#### Vantagens e desvantagens

A técnica de normalização min-max apresenta algumas vantagens e desvantagens que podem influenciar sua aplicação em análises de dados. A seguir, são apresentadas essas vantagens e desvantagens:

##### Vantagens:

- Simplicidade: a normalização min-max é uma técnica simples e fácil de implementar. A fórmula de cálculo envolve apenas algumas operações matemáticas básicas.
- Preserva a relação de ordem: a normalização min-max preserva a ordem dos valores originais. Ou seja, se um valor A é maior que um valor B antes da normalização, a relação de ordem entre A e B será mantida após a normalização.
- Mantém a interpretabilidade: ao transformar os valores em uma escala entre 0 e 1, a normalização min-max preserva a interpretabilidade dos dados. Os valores normalizados ainda podem ser facilmente interpretados em relação ao intervalo original.

##### Desvantagens:

- Sensibilidade a outliers: A normalização min-max é sensível a outliers, que são valores extremos ou discrepantes. Se houver outliers na variável, eles podem distorcer a escala de normalização, comprimindo a maioria dos valores em uma faixa estreita.
- Não é robusta a mudanças nos dados: A normalização min-max é sensível a mudanças nos valores mínimos e máximos da variável. Se novos dados forem adicionados ou se os valores mínimos e máximos mudarem, será necessário recalcular a normalização de todos os dados novamente.
- Pode não lidar bem com distribuições não lineares: A normalização min-max pressupõe uma distribuição linear dos dados. Se os dados possuírem uma distribuição não linear, a normalização min-max pode não ser a técnica mais adequada, pois os valores normalizados podem não refletir adequadamente as relações entre os dados.

É importante considerar essas vantagens e desvantagens ao decidir aplicar a normalização min-max em um conjunto de dados específico. Dependendo da natureza dos dados e do contexto da análise, outras técnicas de normalização, como a normalização z-score, podem ser mais apropriadas.

#### Aplicabilidade e casos de uso

A técnica de normalização min-max tem uma ampla aplicabilidade e é frequentemente utilizada em uma variedade de casos de uso na análise de dados. A seguir são apresentados exemplos de sua aplicabilidade e casos de uso:

- Algoritmos de aprendizado de máquina: a normalização min-max é comumente aplicada como parte da preparação de dados para algoritmos de aprendizado de máquina. Ao trazer todas as variáveis para uma escala comparável, a normalização min-max permite que os algoritmos tratem todas as características igualmente, evitando que recursos com escalas maiores dominem o processo de modelagem.

- Redes neurais: a normalização min-max é frequentemente aplicada em conjunto com redes neurais, que são modelos de aprendizado de máquina sensíveis às magnitudes dos valores de entrada. Normalizar os dados de entrada em uma escala entre 0 e 1 ajuda a garantir que todas as variáveis tenham a mesma influência durante o treinamento da rede neural.

- Análise de dados multivariados: a normalização min-max é útil em análises que envolvem a combinação ou comparação de várias variáveis. - Visualização de dados: a normalização min-max é aplicada em casos em que os dados são exibidos em gráficos ou visualizações. Isso permite que os sistemas de recomendação tratem todas as avaliações de forma equitativa e encontrem padrões e similaridades entre os itens, fornecendo recomendações mais precisas aos usuários.

A normalização min-max é amplamente aplicável em diversos contextos e casos de uso, especialmente quando se deseja trazer as variáveis para uma escala comparável e eliminar o viés introduzido por unidades de medida diferentes. Sua utilização é comum em preparação de dados, algoritmos de aprendizado de máquina, análise de dados multivariados, visualização de dados e sistemas de recomendação.

#### Exemplo em python

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados de exemplo
dados = np.array([10, 20, 30, 40, 50])

# Aplicando a normalização min-max
dados_normalizados = (dados - np.min(dados)) / (np.max(dados) - np.min(dados))

# Plotando os dados originais e normalizados
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.plot(dados, 'bo-', label='Dados Originais')
plt.title('Dados Originais')
plt.xlabel('Índice')
plt.ylabel('Valores')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(dados_normalizados, 'ro-', label='Dados Normalizados')
plt.title('Dados Normalizados (Min-Max)')
plt.xlabel('Índice')
plt.ylabel('Valores Normalizados')
plt.legend()

plt.tight_layout()
plt.show()
```

Neste exemplo, a variável "dados" contém um array com os valores originais que desejamos normalizar. A normalização min-max é realizada através da fórmula (x - min) / (max - min), em que "x" é o valor original, "min" é o valor mínimo do array e "max" é o valor máximo do array. Os dados normalizados são armazenados na variável "dados_normalizados".

Em seguida, utilizamos a biblioteca matplotlib para plotar os dados originais e os dados normalizados em um gráfico de linha. A função plt.subplot é usada para criar dois subplots, sendo um para os dados originais e outro para os dados normalizados. Os rótulos dos eixos e títulos dos gráficos são adicionados para facilitar a compreensão. Finalmente, chamamos plt.tight_layout() para garantir que os gráficos não se sobreponham e plt.show() para exibir o gráfico.

Dessa forma, o exemplo ilustra como realizar a normalização min-max em Python e plotar os dados normalizados em relação aos dados originais.

### Normalização z-score (padronização)

A técnica de normalização z-score, também conhecida como padronização, é um método utilizado para transformar os valores de uma variável de forma que tenham média zero e desvio padrão igual a um. Essa técnica é amplamente aplicada na análise de dados para facilitar a comparação e interpretação dos valores em relação à distribuição dos dados.

A fórmula para o cálculo da normalização z-score é a seguinte:

$$Z = \frac{X−μ}{σ}$$

Onde:

- $Z$ é o valor normalizado (z-score) da variável;
- $X$ é o valor original da variável;
- $μ$ é a média dos valores da variável;
- $σ$ é o desvio padrão dos valores da variável.

Para realizar o cálculo, o subtrai-se o valor médio da variável $μ$ do valor original $X$. Em seguida, o resultado é dividido pelo desvio padrão $σ$ da variável.

Dessa forma, o valor normalizado resultante $Z$ terá média zero e desvio padrão igual a um. Essa padronização permite comparar e interpretar os valores em relação à distribuição dos dados, destacando se um valor está acima ou abaixo da média e em que medida.

A normalização z-score é frequentemente utilizada em análises estatísticas, particularmente em testes de hipóteses, análises de regressão e detecção de outliers. Ela permite identificar valores atípicos (outliers) e entender como cada valor se comporta em relação à distribuição dos dados.

Ao aplicar a normalização z-score, os valores normalizados podem ser interpretados em termos de desvios padrão em relação à média. Valores positivos indicam que o valor original está acima da média, enquanto valores negativos indicam que o valor original está abaixo da média.

A normalização z-score é uma técnica estatística que padroniza os valores de uma variável, tornando sua média igual a zero e seu desvio padrão igual a um. É uma ferramenta útil na análise de dados para comparação e interpretação dos valores em relação à distribuição dos dados.

#### Vantagens e desvantagens

A técnica de normalização z-score, também conhecida como padronização, possui diversas vantagens e algumas desvantagens. A seguir, são apresentadas as vantagens e desvantagens da técnica de normalização z-score:

##### Vantagens

- Comparação direta: a normalização z-score permite a comparação direta dos valores em relação à média e ao desvio padrão dos dados. Isso possibilita identificar valores que estão acima ou abaixo da média e entender seu comportamento em relação à distribuição dos dados.

- Detecção de outliers: a normalização z-score facilita a detecção de outliers, que são valores extremos que se desviam significativamente da média. Valores com z-score alto (positivo ou negativo) são considerados atípicos e podem ser identificados como outliers.

- Equivalência estatística: após a normalização z-score, os valores normalizados têm média zero e desvio padrão igual a um. Isso significa que os valores normalizados estão em uma escala padronizada, permitindo que sejam tratados de maneira equivalente em termos estatísticos.

##### Desvantagens:

- Sensibilidade a outliers: a normalização z-score é sensível a outliers, pois o cálculo do desvio padrão é afetado por valores extremos. Se houver outliers significativos nos dados, eles podem distorcer o cálculo do desvio padrão e afetar a normalização z-score.
- Redução da interpretabilidade: ao realizar a normalização z-score, os valores originais perdem a sua interpretação direta, uma vez que são transformados em desvios padrão em relação à média. Isso pode dificultar a interpretação dos valores normalizados, especialmente em contextos em que a escala original é importante.
- Assunção de normalidade: a normalização z-score assume que os dados seguem uma distribuição normal. Se os dados não seguem essa distribuição, a normalização z-score pode não ser apropriada e resultar em valores normalizados que não refletem corretamente o comportamento dos dados.

É importante considerar essas vantagens e desvantagens ao aplicar a normalização z-score em um conjunto de dados específico. Embora a técnica seja amplamente utilizada e tenha benefícios significativos, é essencial avaliar se é apropriada para o contexto e as características dos dados em questão.

#### Exemplo em python

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados de exemplo com um outlier
dados = np.array([10, 20, 30, 40, 50, 200])

# Cálculo da normalização z-score
z_scores = (dados - np.mean(dados)) / np.std(dados)

# Plotagem dos dados originais e normalizados
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.plot(dados, 'bo-', label='Dados Originais')
plt.title('Dados Originais')
plt.xlabel('Índice')
plt.ylabel('Valores')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(z_scores, 'ro-', label='Dados Normalizados (Z-Score)')
plt.axhline(y=3, color='gray', linestyle='--', label='Limite Superior')
plt.axhline(y=-3, color='gray', linestyle='--', label='Limite Inferior')
plt.title('Dados Normalizados (Z-Score)')
plt.xlabel('Índice')
plt.ylabel('Valores Normalizados')
plt.legend()

plt.tight_layout()
plt.show()
```

Neste exemplo, a variável "dados" contém um array com os valores originais, incluindo um outlier com o valor 200. A normalização z-score é calculada utilizando a fórmula (x - média) / desvio padrão, em que "x" é o valor original, "média" é a média dos dados e "desvio padrão" é o desvio padrão dos dados. ## Seleção da técnica de normalização

### Considerações sobre a distribuição dos dados

A distribuição dos dados é um fator importante a ser considerado na seleção de técnicas de normalização dos dados. Diferentes técnicas de normalização são mais adequadas para diferentes distribuições dos dados, e é essencial escolher a técnica apropriada para preservar a estrutura e as características dos dados.

Ao analisar a distribuição dos dados, alguns aspectos a serem considerados são a presença de assimetria, caudas longas, outliers e a escala dos dados.

Se os dados seguem uma distribuição normal ou aproximadamente normal, a normalização z-score é uma escolha comum. Essa técnica é eficaz para dados que possuem média zero e desvio padrão um, o que facilita a interpretação dos valores em termos de desvios em relação à média. Além disso, a normalização z-score é robusta em relação a outliers moderados, pois utiliza a média e o desvio padrão como referência.

No entanto, se os dados não seguem uma distribuição normal ou possuem caudas longas, a normalização z-score pode distorcer a distribuição e não ser a melhor opção. Nesses casos, a normalização min-max pode ser mais adequada. Essa técnica redimensiona os valores para um intervalo específico, como 0 e 1, preservando a ordem relativa dos dados. A normalização min-max é útil quando é importante manter a escala dos dados e quando não há suposições fortes sobre a distribuição subjacente.

Outra consideração importante é a presença de outliers. A normalização z-score pode ser sensível a outliers extremos, uma vez que utiliza a média e o desvio padrão, que são afetados por valores atípicos. Nesses casos, técnicas de normalização robustas a outliers, como a normalização por mediana e intervalo interquartil (IQR), podem ser mais apropriadas.

Além disso, é importante lembrar que a normalização dos dados não é sempre necessária ou adequada. Em alguns casos, as técnicas de normalização podem não ser aplicáveis ou podem até mesmo introduzir viés ou distorção nos dados. É fundamental compreender a natureza dos dados, o objetivo da análise e as implicações de diferentes técnicas de normalização antes de aplicá-las.

Considerar a distribuição dos dados é fundamental ao selecionar técnicas de normalização. É importante escolher a técnica que preserva a estrutura e as características dos dados, levando em conta a distribuição, a presença de outliers e a escala dos dados. A seleção adequada da técnica de normalização contribui para análises mais precisas e insights valiosos a partir dos dados.

## Normalização dos dados de teste ou novos dados

A normalização dos dados de teste ou novos dados é uma etapa importante para garantir a consistência e a precisão das análises e modelos. Uma prática comum é utilizar os mesmos parâmetros de normalização obtidos a partir dos dados de treinamento nos dados de teste ou novos dados.

Os parâmetros de normalização referem-se aos valores utilizados para normalizar os dados, dependendo da técnica escolhida. Eles são calculados a partir dos dados de treinamento e são usados para padronizar os dados de teste ou novos dados da mesma maneira.

Quando um modelo é treinado usando dados normalizados, ele aprende os padrões e relações entre as variáveis nessa escala específica. Portanto, é fundamental aplicar a mesma transformação de normalização nos dados de teste ou novos dados antes de fazer previsões ou realizar análises.

Ao usar os mesmos parâmetros de normalização, como média e desvio padrão para a normalização z-score ou valores mínimo e máximo para a normalização min-max, garante-se que os dados de teste ou novos dados estejam na mesma escala que o modelo foi treinado. Isso permite que o modelo interprete corretamente os valores dos dados e produza resultados consistentes e confiáveis.

A utilização dos mesmos parâmetros de normalização também é importante para evitar a introdução de viés nos resultados. Se diferentes parâmetros de normalização forem aplicados nos dados de treinamento e de teste, os dados estarão em escalas diferentes e as relações entre as variáveis podem ser distorcidas. Isso pode levar a previsões imprecisas e análises enviesadas.

Além disso, é importante ressaltar que os parâmetros de normalização obtidos nos dados de treinamento devem ser armazenados e aplicados consistentemente nos dados de teste ou novos dados. Esses parâmetros são uma parte crucial do processo de normalização e devem ser utilizados toda vez que os dados forem normalizados, mesmo após o treinamento inicial do modelo.

No entanto, é importante lembrar que os parâmetros de normalização obtidos nos dados de treinamento não devem ser atualizados ou recalculados com os dados de teste ou novos dados. Isso garantirá a consistência e a validade das análises e modelos, evitando a introdução de informações adicionais do conjunto de teste nos parâmetros de normalização.

A normalização dos dados de teste ou novos dados utilizando os mesmos parâmetros de normalização obtidos nos dados de treinamento é uma prática fundamental para manter a consistência e a precisão das análises e modelos. Isso garante que os dados estejam na mesma escala em que o modelo foi treinado e evita a introdução de viés nos resultados. Os parâmetros de normalização devem ser armazenados e aplicados de forma consistente em todas as etapas de normalização, garantindo a validade das análises e modelos.
